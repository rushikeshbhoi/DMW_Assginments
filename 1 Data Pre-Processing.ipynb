```python
#import libraries
import numpy as numpy
import pandas as pd
```

```python
# read data from dataset into dataframe
df = pd.read_csv('world-population-by-country-2020.csv')
```

```python
# display first 5 rows
df.head()
```

```python
# get information about dataframe
df.info()
```

```python
# data reduction
# remove/reduce unnecessary attributes from dataframe
subset = df[['no','Country (or dependency)','Population 2020','Yearly Change','Net Change','Migrants (net)']]
subset.head()
```

```python
# data cleaning 
# find missing values
subset.isnull().sum()
```

```python
# cleaning the null data
subset.dropna(inplace=True)
subset.isnull().sum()
```

```python
subset2 = df[['Net Change','Yearly Change']]
subset2.head()
```

```python
# data integration
subset3 = pd.merge(subset, subset2, on='Yearly Change')
subset3.head()
```

```python
# data transformation
subset3.to_excel('World_Population.xlsx')
```
